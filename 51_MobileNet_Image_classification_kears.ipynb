# MobilNets
In this episode, we'll introduce MobileNets, a class of light weight deep convolutional neural
networks that are vastly smaller in size and faster in performance than many other popular
models. We'll also see how we can work with MobileNets in code using TensorFlow's Keras API.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
from tensorflow.keras.applications import imagenet_utils
from sklearn.metrics import confusion_matrix
import itertools
import os
import shutil
import random
import matplotlib.pyplot as plt
```

```python
mobile = tf.keras. applications.mobilenet.MobileNet()
```

```python
def prepare_image(file):
  img_path = r'data/'
  img = image.load_img(img_path + file, target_size=(224, 224))
  img_array = image.img_to_array(img)
  img_array_expanded_dims = np.expand_dims(img_array, axis=0)
  return tf.keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)
```

```python
from IPython.display import Image
Image(filename='data/mobilenet1.png', width=450,height=300)
```

```python
preprocessed_image = prepare_image('mobilenet1.png')
predictions = mobile.predict(preprocessed_image)
#predictions
results = imagenet_utils.decode_predictions(predictions)# to return Probability
results
```
