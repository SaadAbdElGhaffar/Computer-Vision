```python
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/qrcode.png")

decodedObjects = pyzbar.decode(image)
for obj in decodedObjects:
  print(obj)
  print("==================")
  print("Type:", obj.type)
  print("Data: ", obj.data, "\n")

image = cv2_imshow(image)

cv2.waitKey(0)
```
