```python
import pydotplus
from IPython.display import display, Image
import matplotlib.pyplot as plt
import pytesseract
from gtts import gTTS
import IPython.display as ipd # we will used display to display voice
from requests import get # to make GET request
```

```python
# This cell is reserved for using pydotplus to visualize the plan.
myplan="""digraph {
Importing_the_necessary_Libraries ->
dowenload_arabic_traineddata ->
Read_image_by_cv2 ->
Generate_OCR_Text_by_pytesseract ->
Read_OCR_Text_by_gTTS
}"""

mygraph=pydotplus.graph_from_dot_data(myplan)
mygraph.write_png("myplan.png")
display(Image(filename='./myplan.png'))
```

```python
def download(url, file_name):
  # open in binary mode
  with open(file_name, "wb") as file:
    # get request
    response = get(url)
    # write to file
    file.write(response.content)

download("https://github.com/tesseract-ocr/tessdata/raw/master/ara.traineddata","pytesseract/Tesseract-OCR/tessdata/ara.traineddata")
```

```python
import cv2

filename="data/Arabic_OCR.png"
img_cv = cv2.imread(filename)
plt.figure(figsize=(20, 20))
plt.imshow(img_cv,cmap = 'gray')
plt.title('template'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
pytesseract.pytesseract.tesseract_cmd = r"pytesseract/Tesseract-OCR/tesseract.exe"
arabic_text=pytesseract.image_to_string(img_cv , lang='ara',config= ".")

print(arabic_text)
```

```python
tts = gTTS(arabic_text, lang='ar')
tts.save('output/test.mp3')
```

```python
audio_path="output/test.mp3"

ipd.Audio(audio_path, autoplay=True)
```
